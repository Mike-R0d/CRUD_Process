```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Create a 2x2 subplot layout
fig = make_subplots(rows=2, cols=2,
                    subplot_titles=("Quebec", "Alberta", "Ontario", "British Columbia"),
                    specs=[[{"secondary_y": True}, {"secondary_y": True}],
                           [{"secondary_y": True}, {"secondary_y": True}]])

# Add the primary traces (actual values)
fig.add_trace(go.Scatter(x=table.index, y=table['Quebec'], name='Quebec'), row=1, col=1, secondary_y=True)
fig.add_trace(go.Scatter(x=table.index, y=table['Alberta'], name='Alberta'), row=1, col=2, secondary_y=True)
fig.add_trace(go.Scatter(x=table.index, y=table['Ontario'], name='Ontario'), row=2, col=1, secondary_y=True)
fig.add_trace(go.Scatter(x=table.index, y=table['British Columbia'], name='British Columbia'), row=2, col=2, secondary_y=True)

# Add the percentage variation traces on the secondary y-axes
fig.add_trace(go.Bar(x=table.index, y=table['Quebec'].pct_change() * 100, name='Quebec % Change'), row=1, col=1, secondary_y=False)
fig.add_trace(go.Bar(x=table.index, y=table['Alberta'].pct_change() * 100, name='Alberta % Change'), row=1, col=2, secondary_y=False)
fig.add_trace(go.Bar(x=table.index, y=table['Ontario'].pct_change() * 100, name='Ontario % Change', ), row=2, col=1, secondary_y=False)
fig.add_trace(go.Bar(x=table.index, y=table['British Columbia'].pct_change() * 100, name='British Columbia % Change', ), row=2, col=2, secondary_y=False)

# Update layout to set the title and size
fig.update_layout(height=800, width=1200, title_text="Provincial GDP in Canada with Percentage Variation")

fig.update_layout(
    xaxis_showgrid=False,  # Activate vertical grid lines
    yaxis_showgrid=False,  # Activate horizontal grid lines
    template = "plotly_dark",
)

# Show the plot
fig.show()
```

In this stage we are creating tables on SQLite




```python
import sqlite3
import pandas as pd

path_db = r'C:\Users\LENOVO\Documents\my_database.db'
table_name = 'pib_privinces'

# Connect to the SQLite database
sqlite_connection = sqlite3.connect(path_db)
sqlite_cursor = sqlite_connection.cursor()

# Get the DataFrame headers to use them as column names
columns = df.columns

# Create the query to create the table using the DataFrame's column names
create_table_query = f"CREATE TABLE IF NOT EXISTS {table_name} ({' TEXT, '.join(columns)} TEXT)"

# Execute the query to create the table
sqlite_cursor.execute(create_table_query)

# Confirm the changes
sqlite_connection.commit()

# Insert the DataFrame data into the specified table
data_to_insert = df.to_records(index=False)
columns_to_insert = ', '.join(columns)
placeholders = ', '.join(['?'] * len(columns))

sqlite_cursor.executemany(f"INSERT INTO {table_name} ({columns_to_insert}) VALUES ({placeholders})", data_to_insert)

sqlite_connection.commit()
sqlite_connection.close()

print(f"Data successfully inserted into the {table_name} table.")
```

```python
import sqlite3
import pandas as pd

# Path to the SQLite database
path_db = r'C:\Users\LENOVO\Documents\my_database.db'

# Connect to the SQLite database
sqlite_connection = sqlite3.connect(path_db)
cursor_sqlite = sqlite_connection.cursor()

# Retrieve the data from the 'pib_provinces' table
query = 'SELECT * FROM pib_provinces'
cursor_sqlite.execute(query)

# Fetch all rows
pib_sqlite = cursor_sqlite.fetchall()

# Get the column names from the cursor
columns = [description[0] for description in cursor_sqlite.description]

# Convert the fetched data into a DataFrame
df_pib_provinces = pd.DataFrame(pib_sqlite, columns=columns)

# Close the connection
sqlite_connection.close()
df_pib_provinces
```
